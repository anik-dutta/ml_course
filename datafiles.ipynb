```python
import pandas as pd
emp=pd.read_csv("Employee.csv")
empdf=pd.DataFrame(emp)
print(empdf)
print('\n')
dept=pd.read_csv('Department.csv')
deptdf=pd.DataFrame(dept)
print(deptdf)
```

```python
empdf.info()
empdf['sal'].describe()
empdf.head()
x=empdf[['ename','job','mgr']]
print(x)
print(x.ndim)
```

```python
#1 row filtering
tempdf=empdf.loc[empdf["deptno"]==30]
print(tempdf)
```

```python
#2
tempdf1=empdf.loc[(empdf["sal"]>700) & (empdf["deptno"]==30)]
print(tempdf1[["ename","sal","deptno"]])
```

```python
arr=[10,20]
tempdf2=empdf.loc[empdf["deptno"].isin(arr)]
print(tempdf2[["ename","sal","deptno"]])
```

```python
import numpy as np
grpemp=empdf.groupby(["deptno"])
for d,group in grpemp: #d is the keyword grpemp is the group
    print("the dept is:",d)
    print("the group is:",group)
    print("the name is:",group["ename"])
cnt=empdf.groupby(["deptno"])
print("the count is:",cnt["deptno"].agg(np.count_nonzero))
print("the sum is:",cnt["sal"].agg(np.sum))
```
