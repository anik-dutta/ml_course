```python
#string function
str="helloworld"
if "a" in str:
    print("char a is present in str")
else:
    print("char a is not present in str")

if "a" not in str:
    print("char a is not present in str")
else:
    print("char a is present in str")
```

```python
#in built string functions
str="helloworld"
#iength of string
print("length is :",len(str))

#remove all leading & trailing spaces
print("string after stripping:",str.strip())
print("strin length after stripping:",len(str.strip()))

#split string at specified char
str="hello,world,all"
print("string after split:",str.split(","))
#max split
print("string after max split:",str.split(",",1))

#counts the occurence of one string in another
str1="hello world"
print("no of occurences:",str1.count("l"))
print("no of occurences:",str1.count("l",0,5))

#find
str2="helloworld"
print("found or not",str2.find("l",0,len(str2)))
#is digit
str3="4"
print("is digit or not",str3.isdigit())
```

```python
#creating list
list1=[1,5,3,7,5]
print("the list:",list1)
print("the length of list:",len(list1))
print("the value of elements 2 to 4 :",list1[1:4])
for x in list1:
    print("the values in the list are:",x)
    
#updating list
list1[2]=21
print("list after updation:",list1)
```

```python
#insert value
list1=[1,3,5,7,5]
list1.insert(2,101)
print("list after addind at 2nd index:",list1)
#append a value
list1=[1,3,5,7,5]
list1.append(102)
print("the length after adding:",len(list1))
print("list after adding:",list1)

#adding list inside a list
list1.append([110,111])
print("length after adding:",len(list1))
print("list after adding:",list1)

#adding elements in list through extend method
list1=[1,5,3,7,5]
list1.extend([110,111])
print("length after adding:",len(list1))
print("list after adding:",list1)
```

```python
#delete from index
list1=[1,3,5,7,5]
del list1[0]
print("list after deletion:",list1)
#delete exact value from list
list1=[1,3,5,7,5]
list1.remove(7)
print("list is :",list1)
#removing & returning by pop method
list1=[1,3,5,7,5]
var=list1.pop(0)
print("list:",list1)
print("var :",var)
```

```python
#sorting a list
list1=[1,3,5,7,5]
print(list1)
list1.sort()
print("In ascending order the list is:",list1)
list1.sort(reverse=True)
print("In descending order the list is:",list1)
```

```python
#take all the strings more than of length 4 and having 1st and last char same
a=["madam","xyzyx","ma'am","abcda"]
c=0
for r in a:
     if len(r)>4 and r[0]==r[len(r)-1]:
        c+=1
print(c)
```

```python
#delete all the even nums
a,p=[2,3,4,2,5,9],[]
for i in a:
    if i%2!=0:
        p.append(i)
print(p)
```

```python
#delete the duplicates
x,y=[1,2,3,2,1],[]
for i in x:
    if i not in y:
        y.append(i)
print(y)
```
