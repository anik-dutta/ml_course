```python
#prime number
flag=0
for i in range (1,20):
    if i==1 or i==0:
        continue
        flag=1
    for j in range (2,i):
        if i%j==0:
            flag=0
            break
    if flag==1:
        print(i)
```

```python
import math
c=9.7
#var='a'

print("the ceiling value is :",math.ceil(c))
print("the floor value is :",math.floor(c))
print("the sqrt value is :",math.sqrt(4))
print("the factorial value is :",math.factorial(3))

#naive python mathematical functions
print("The max value is :",max(5,3,6))
print("The min value is :",min(5,3,6))
print("The round value is :",round(5.4))
```

```python
#random numbers
import random
#generates floating numbers between 0 & 1
print("random number:",random.random())

#generates random nums within a range from excluding upperbound
print("1.Range from list:",random.randrange(999))

#generates random nums within a range from base value to range skipp
print("2.range from list :",random.randrange(2,12,4))

#randint()
#generates random nums between two int including both
x=0
while x<10:
    print("3.Random numbers :",random.randint(0,9))
    x=x+1
    
#choice()
#generates a random value from list
print("4.the value is :",random.choice([1,5,3]))
```

```python
#string slicling
str="hellobae"
print("1st char :",str[0])
print("last char :",str[-1])
print("0 to 4th char :",str[:4])
print("2nd to 5th char :",str[1:5])
print("3rd to 3rd char : ",str[2:-2])

#iterating through strings
str="hellobwjb"
for ch in str:
    print(ch)
```

```python
a="hello"
n=5
for i in range (n):
    for j in range (n-i+1):
        print("",end=" ")
    for k in range (0,i+1):
        print(a[k:k+1],end=" ")
    print()
for i in range (1,n):
    print(" ",end=" ")
    for j in range (i):
        print("",end=" ")
    for l in range(0,n-i):
        print(a[l:l+1],end=" ")
    print()
```
