```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('Headbrain.csv')
#z=data.head()
x1=np.mean(data['Head size'])
print(x1)
y1=np.mean(data['Brain Weights'])
print(y1)
sum1=0
sum2=0
x=data['Head size']
y=data['Brain Weights']
size=len(data['Head size'])
for  i in range(size):
    sum1=sum1+(x[i]-x1)*(y[i]-y1)
    sum2=sum2+(x[i]-x1)**2
m=sum1/sum2
print(m)
c=y1-m*x1
print(c)
lis=[]
for z in range(size):
        yp=np.array(m*x[z]+c)
        lis.append(yp)
data['predict y']=lis
print(data)
nu=0
de=0
for z in range(size):
    nu=nu+((lis[z]-y1)**2)
    de=de+((y[z]-y1)**2)
r=nu/de
print(r)        
```
