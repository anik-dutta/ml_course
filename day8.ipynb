```python
import numpy as np
import pandas as pd
df=pd.DataFrame(np.random.randn(5,3),index=['a','c','e','f','h'],columns=['one','two','three'])
print(df)
#reindexing
df=df.reindex(['a','b','c','d','e','f','g','h'])
print(df)
print("\n")
#count the no. of NaN in a column
print(df['one'].isnull().sum())
print("\n")
#sum of a column
print(df[['one','two']].sum())
print("\n")
print(df)
#cleaning data
print(df['one'].fillna(np.mean(df['one'])))
print("\n")
print(df.fillna(method="ffill"))
print("\n")
print(df.fillna(method="bfill"))
```

```python
#Bar graph

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
company=["IBM","TCS","CTS","INFOSYS"]
revenue=[10,89,40,56]
profit=[5,20,12,10]
xpos=np.arange(len(company))
#xticks is list of positions where ticks need to be given
plt.xticks(xpos,company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company Revenue")
#plt.bar(xpos,revenue,label="Revenue") #for a merged bar
#plt.bar(xpos,profit,label="Profit")   #for a merged bar
plt.bar(xpos-0.2,revenue,width=0.3,label="Revenue",color='purple')    #for an individual bar
plt.bar(xpos+0.2,profit,width=0.3,label="Profit",color='red')     #for an individual bar
plt.legend()
plt.show()
```

```python
plt.yticks(xpos,company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company revenue")
#plt.barh(xpos,revenue,label="Revenue") 
#plt.barh(xpos,profit,label="Profit")   
plt.barh(xpos+0.15,profit,height=0.2,label="Profit")
plt.barh(xpos-0.15,revenue,height=0.2,label="Revenue")
plt.legend()
plt.show()
```

```python
#Piechart

expval=[5000,500,2000,800]
exphead=['House Rent','Mobile Bill','Food','Trip']
plt.axis()
plt.pie(expval,labels=exphead,radius=1.5,autopct='%0.5f%%',shadow=True,explode=[0,0,0,0.2])
plt.show()
```
