```python
import numpy as np
import pandas as pd
df=pd.read_csv('Purchase.csv')
cdf=pd.DataFrame(df)
cdf.head(15)
```

```python
#separate the data
X=df.drop(['User ID','Purchased'],axis=1)
Y=df['Purchased']

#convert labels to numbers for input variable
from sklearn.preprocessing import LabelEncoder
le_Gender=LabelEncoder()
#add the le_Gender columns in input var df
X['Gender_n']=le_Gender.fit_transform(X['Gender'])
print(X.head())
```

```python
#drop the original gender column
X.drop('Gender',axis=1,inplace=True)
```

```python
#spliting data in train & test set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=4)
```

```python
#function to perform training using entropy
from sklearn.tree import DecisionTreeClassifier
classifier_entropy=DecisionTreeClassifier(criterion='entropy',random_state=100,max_depth=3)
#create the model
classifier_entropy.fit(X_train,Y_train)
```

```python
#make prediction
y_pred=classifier_entropy.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_pred))
```

```python
#check accuracy
from sklearn.metrics import accuracy_score
print("Accuracy is:",accuracy_score(Y_test,y_pred)*100)
```
