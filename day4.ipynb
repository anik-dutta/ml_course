```python
#dictionary
dict1={
        "st":"sachin tendulkar",
        "msd":"mahendra singh dhoni",
        "runs":10000  }
print(dict1)
print(dict1["runs"])
for x in dict1:
    print(x)
for x in dict1:
    print(dict1[x])
for x in dict1:
    print(x,":",dict1[x])
```

```python
#inserting item in dictionary
dict1["sv"]="sourav ganguly"
print(dict1)
```

```python
#updating value in dictionary
dict1["runs"]=34500
print(dict1)
print(dict1["runs"])
print(len(dict1))
```

```python
#remove items from dict
dict1={
        "st":"sachin tendulkar",
        "msd":"mahendra singh dhoni",
        "runs":10000  }
print(dict1)
var=dict1.pop("runs")
print(dict1)
print(var)
var1=dict1.popitem()
print(dict1)
print(var)
```

```python
#creating a tuple
tuple1=("sachin","msd","sourav")
print(tuple1[0],tuple1[1])
for x in tuple1:
    print(x)
```

```python
#create a tuple or tuple packing
tuplex=4,8,4
print(tuplex)
n1,n2,n3=tuplex
#UNPACK A TUPLE IN VARIABLES
print(n1+n2+n3)
```

```python
#creating a set
days=set(["mon","tue","wed","thu","fri","sat","sun","sun"])
months={"jan","feb","mar"}
dates={21,22,3}
print(days)
print(months)
print(dates)
print(type(dates))
```

```python
#adding items to a set
#set is immutable here
days=set(["mon","tue","wed","thu","fri","sat"])
print(id(days))
days.add("abs")
print(id(days))
print(days)
days={"cde"}
print(id(days))
```

```python
#removing from a set
print(id(days))
days=set(["mon","tue","wed","thu","fri","sat"])
days.discard("sun")
print(days)
```

```python
#union of set
daysA=set(["wed","sun"])
daysB=set(["mon","tue","wed","thu","fri","sat"])
alldays=daysA|daysB
print(alldays)
#intersection of set
alldays=daysA & daysB
print(alldays)
```

```python
import numpy as np
#1D array
arr1=np.array([1,2,3])
print(arr1)
print(type(arr1))
print(arr1.dtype)
print(arr1[0])
```

```python
#nD array
arr2=np.array([[1,2,3],[4,5,6]])
print(arr2)
print(arr2[1,1],arr2[0,2])
```

```python
arr2=np.array([(9),(2)])
print(arr2)
print(arr2.ndim)
arr2=np.array([(9,),(2,)])
print(arr2)
print(arr2.ndim)
```

```python
import numpy as np
#data tyoe
arr3=np.array([9,2],dtype=float)
print(arr3)
print(arr3.dtype)
```

```python
#shape returns tuple consisting of array dimensions
arr1=np.array([[10,20],[30,40],[30,40]])
print("array shape is :",arr1.shape)
print("array is:",arr1)
```

```python
#resize the shape
arr1=np.array([[10,20,30],[30,40,50]])
print("array shape is :",arr1.shape)
arr1.shape=(3,2)
print("array shape is :",arr1.shape)
print("array is:",arr1)
```

```python
#resizing through reshape()
arr2=np.array([[10,23,30,40],[30,40,50,60]])
print(arr2)
b=arr2.reshape(4,2)
print("array shape is:",arr2.shape)
print("array is :",b)
```

```python
#array creation routines
#empty() creates an uninitialized array
arr1=np.array([3,2],dtype=int)
print("filled with random values:",arr1) #filled with random values

#Zeros creates an array filled with zeroes
arr2=np.zeros([3,2])
print("filled with 0 ints:",arr2)

#ones 
arr3=np.ones([3,2])
print("filled with one floats:",arr3)
```

```python
#filling nd array from existing data structures
list1=[1,2,3,4]
arr1=np.array(list1)
print(type(list1))
print(list1)
```

```python
#ndarray from tuple
tup1=(1,2,3)
arr3=np.array(tup1)
print(arr3)

#ndarray from a list of tuple
list1=[(1,2,3),(4,5,6)]
arr4=np.array(list1)
print("dim of array:",arr4.ndim)
print("array from list of tuples:",arr4)
```

```python
#fromiter--ndarray from iterator object
list1=range(5)
x=iter(list1)#returns iterator object

arr5=np.fromiter(x,dtype=float)
print("array from iterator object:",arr5)
```
