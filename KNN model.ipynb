```python
#KNN predict whether a person has diabetes or not

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('Diabetes.csv')
df.head()
```

```python
df.info()
```

```python
#removing the zeros

zero_not_accepted=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for column in zero_not_accepted:
    df[column]=df[column].replace(0,np.NaN)
    mean=int(df[column].mean(skipna=True))  #skipna=True is used for not considering the NaN values
    df[column]=df[column].replace(np.NaN,mean)
    print(df[column])
```

```python
#split dataset into train and test
X=df.iloc[:,0:8]
Y=df.iloc[:,8]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=4,test_size=0.3)
```

```python
#feature scaling

sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.fit_transform(X_test)
print(X_train)
```

```python
#define the model
#import math
#print(math.sqrt(len(X_test)))
classifier=KNeighborsClassifier(n_neighbors=19,p=2,metric='euclidean')
classifier.fit(X_train,Y_train)
```

```python
#predict the test result set

y_pred=classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
print(cm)
```
