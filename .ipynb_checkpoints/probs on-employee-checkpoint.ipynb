```python
import pandas as pd
emp=pd.read_csv("Employee.csv")
empdf=pd.DataFrame(emp)
print(empdf)
print('\n')
dept=pd.read_csv('Department.csv')
deptdf=pd.DataFrame(dept)
print(deptdf)
tempdf=empdf.loc[empdf["job"]=="CLERKS"]
print(tempdf)
```

```python
#1.List the employee name,employee no. & dept no. of all clerks

tempdf=empdf.loc[empdf["job"]=="CLERK"]
print(tempdf[["ename","empno","deptno"]])
```

```python
#2.Find the dept no. & employee name whose dept no. is greater than 20

tempdf=empdf.loc[empdf["deptno"]==30]
print(tempdf[["ename","deptno"]])
```

```python
#3.Find the employee whose commission is greater than 30% of their salaries

tempdf=empdf.loc[empdf["comm"]>(empdf["sal"]*0.3)]
print(tempdf[["ename","empno","deptno"]])
```

```python
#4.Find the employee name,job,salary who work(s) in dept no. 20 & who earn(s) more than 2000

tempdf=empdf.loc[(empdf["deptno"]==20) & (empdf["sal"]>2000)]
print(tempdf[["ename","job","sal"]])
```

```python
#5.Find all salesman in dept no. 30 whose salary is greater than 1500

tempdf=empdf.loc[(empdf["deptno"]==30) & (empdf["sal"]>1500) & (empdf["job"]=="SALESMAN")]
print(tempdf[["ename","empno","sal"]])
```

```python
#6.Find all employee names whose designation is either manager or president

tempdf=empdf.loc[(empdf["job"]=="MANAGER")|(empdf["job"]=="PRESIDENT")]
print(tempdf[["ename","job","empno","deptno"]])
```

```python
#7.Find all managers who are not in dept no. 30

tempdf=empdf.loc[(empdf["job"]=="MANAGER") & (empdf["deptno"]!=30)]
print(tempdf[["ename","job","empno","deptno"]])
```

```python
#8.Find the details of all managers(in any dept) & clerks in dept no 20

tempdf=empdf.loc[((empdf["deptno"]==20) & (empdf["job"]=="CLERK")) | (empdf["job"]=="MANAGER")]
print(tempdf[["ename","empno","sal","job"]])
```

```python
#9.Find the employees who don't receive commission

tempdf=empdf.loc[(empdf["comm"]==0.0)|(empdf["comm"]=="NaN")]
print(tempdf[["ename","empno","comm"]])
```

```python
#10.Display the highest,lowest,sum,average salary for each job

import numpy as np
grpemp=empdf.groupby(["job"])
for j,group in grpemp:
    print("\nthe job is:",j)
    print("\nthe group is:\n",group)
    print("\nthe names are:\n",group["ename"])
cnt=empdf.groupby(["job"])
print("the count is:",cnt["job"].agg(np.count_nonzero))
print("\nThe highest is:",cnt["sal"].agg(np.max))
print("\nThe lowest is:",cnt["sal"].agg(np.min))
print("\nThe sum is:",cnt["sal"].agg(np.sum))
print("\nThe average is:",cnt["sal"].agg(np.average))
```
