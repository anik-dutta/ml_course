```python
import seaborn as sns
import numpy as np
import pandas as pd
df=pd.read_csv("cancer.csv")
df=df.drop(['id','Unnamed: 32'],axis=1)
print(df)
for i in df:
    print(i)
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=4)
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
#Confusion Matrix

from sklearn.metrics import confusion_matrix
res=confusion_matrix(y_test,predictions)
res
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,16))
sns.heatmap(df.corr(),annot=True,linewidths=0.5,fmt='.2f')
```

```python
#Train Data
from sklearn.model_selection import train_test_split
feature=['symmetry_worst','area_worst','symmetry_mean','texture_worst','area_mean','smoothness_worst','texture_mean','smoothness_mean','concave points_mean','concave points_worst','concavity_mean','concavity_worst','texture_se','area_se','smoothness_se']
x=df[feature]
y=df.diagnosis
#from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
res=confusion_matrix(y_test,predictions)
res
```

```python
f_drop=['radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst','concavity_se','perimeter_se','concavity_mean','area_mean','compactness_se','concave points_mean']
df=df.drop(f_drop,axis=1) #f_drop: feature drop
```

```python
plt.figure(figsize=(20,16))
sns.heatmap(df.corr(),annot=True,linewidths=0.5,fmt='.2f')
```

```python
#Train Data
from sklearn.model_selection import train_test_split
x=df.drop('diagnosis',axis=1)
y=df.diagnosis
#from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
res=confusion_matrix(y_test,predictions)
res
```

```python
class_names=[1,0]
fig,ax=plt.subplots()
tick_marks=np.arange(len(class_names))
plt.xticks(tick_marks,class_names)
plt.yticks(tick_marks,class_names)
sns.heatmap(pd.DataFrame(res),annot=True,fmt='d')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('confusion',y=1.1)
plt.ylabel('actual label')
plt.xlabel('predicted label')
```

```python
tn,fp,fn,tp=confusion_matrix(y_test,predictions).ravel()
print(tn)
print(fp)
print(fn)
print(tp)
Accuracy=(tn+tp)*100/(tp+tn+fp+fn)
print("Accuracy",Accuracy)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
print("accuracy:",accuracy_score(y_test,predictions)*100)
x1=(classification_report(y_test,predictions))
print(x1)
```
