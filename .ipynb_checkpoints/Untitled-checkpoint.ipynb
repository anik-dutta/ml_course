```python
#selectkbest
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_classif
```

```python
x=df.drop("diagnosis",axis=1)
y=df['diagnosis']
```

```python
bestfeatures=SelectKBest(score_func=f_classif,k=5)
fit=bestfeatures.fit(x,y)
```

```python
dfscores=pd.DataFrame(fit.scores_)
dfcolumns=pd.DataFrame(x.columns)
featurescores=pd.concat([dfcolumns,dfscores],axis=1)
featurescores.columns=['Specs','Scores']
```

```python
print(featurescores)
```

```python
print(featurescores.nlargest(15,'Scores'))
```

```python
#Decision Tree
f_drop=['radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst','concavity_se','perimeter_se','concavity_mean','area_mean','compactness_se','concave points_mean']
df=df.drop(f_drop,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le_diagnosis=LabelEncoder()
df['Diagnosis']=le_diagnosis.fit_transform(df['diagnosis'])
df.drop('diagnosis',axis=1,inplace=True)
print(df.head(25))
```

```python
ndf=pd.DataFrame()
```

```python
X=df.drop(['Diagnosis'],axis=1)
Y=df['Diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=4)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier_entropy=DecisionTreeClassifier(criterion='entropy',random_state=42,max_depth=3)
#create the model
classifier_entropy.fit(X_train,Y_train)
```

```python
y_pred=classifier_entropy.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_pred))
```

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
iqr=q3-q1
print(iqr)
ub=q3+1.5*iqr
lb=q1-1.5*iqr

median=np.median(df['radius_mean'])
print(median)
```

```python
z=['radius_mean','area_mean']
for i in z:
    df[i]=df[i].replace(0,np.NaN)
    mean=int(df[i].mean(skipna=True))  #skipna=True is used for not considering the NaN values
    df[i]=df[i].replace(np.NaN,mean)
    print(df[i])
```

```python
c=0
r=len(df['radius_mean'])
for j in range (r):
    if(df.iloc[j]['radius_mean']>ub) | (df.iloc[j]['radius_mean']<lb):
        c=c+1
print((c/r)*100)
```
