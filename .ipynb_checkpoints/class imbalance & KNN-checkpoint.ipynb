```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Cancer.csv')
df.head()
```

```python
f_drop=['radius_worst','radius_mean','radius_se','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst','concavity_se','perimeter_se','concavity_mean','area_mean','concavity_se','compactness_se','concave points_mean']
df=df.drop(f_drop,axis=1) #f_drop: feature drop
```

```python
df.drop(['Unnamed: 32','id',],inplace=True,axis=1)
print(df.head())
```

```python
import seaborn as sns
sns.countplot(df['diagnosis'])
```

```python
zero_not_accepted=['texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean',\
                  'concave points_mean','texture_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst',\
                  'concave points_worst','symmetry_worst','fractal_dimension_worst']
for column in zero_not_accepted:
    df[column]=df[column].replace(0,np.NaN)
    mean=int(df[column].mean(skipna=True))
    df[column]=df[column].replace(np.NaN,mean)
    print(df[column])
```

```python
import seaborn as sns
sns.countplot(df['diagnosis'])
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make=LabelEncoder()
df["diagnosis_code"]=lb_make.fit_transform(df["diagnosis"])
df[["diagnosis_code","diagnosis"]].head(21)
```

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis']=[1 if b=='B' else 0 for b in df.diagnosis]
df['diagnosis'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import seaborn as sns
sns.countplot(df['diagnosis'])
```

```python
y=df.diagnosis
x=df.drop('diagnosis',axis=1)
clf_0=LogisticRegression().fit(x,y)
pred_y_0=clf_0.predict(x)
```

```python
print(accuracy_score(pred_y_0,y))
```

```python
print(np.unique(pred_y_0))
```

```python
from sklearn.utils import resample
```

```python
df_majority=df[df.diagnosis==0]
df_minority=df[df.diagnosis==1]
```

```python
df_minority_upsample=resample(df_minority,replace=True,n_samples=357,random_state=4)
```

```python
df_upsample=pd.concat([df_majority,df_minority_upsample])
df_upsample.diagnosis.value_counts()
```

```python
y=df.diagnosis
x=df.drop('diagnosis',axis=1)
clf_1=LogisticRegression().fit(x,y)
pred_y_1=clf_1.predict(x)
print(np.unique(pred_y_1))
print(accuracy_score(y,pred_y_1))
```

```python
#KNN model
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
#split dataset and into train and test
X=df.iloc[:,0:30]
Y=(df['diagnosis'])
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=4,test_size=0.3)
```

```python
#Define the model:Init
classifier=KNeighborsClassifier(n_neighbors=5,p=2,metric='euclidean')
classifier.fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_pred))
```
